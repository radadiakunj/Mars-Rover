```python
import cv2

left = cv2.imread('left2.jpg')
right = cv2.imread('right2.jpg')

left_gray = cv2.cvtColor(left, cv2.COLOR_BGR2GRAY)
right_gray = cv2.cvtColor(right, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imwrite('left_gray.jpg', left_gray)
cv2.imwrite('right_gray.jpg', right_gray)

print("Done — left_gray.jpg and right_gray.jpg saved")
```
